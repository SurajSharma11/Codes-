```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
UniversalBank = pd.read_csv("UniversalBank.csv")
```

```python
UniversalBank = UniversalBank.drop(["ID","ZIP Code"],axis = 1)
```

```python
charc_col = ['Family', 'Education','Personal Loan', 'Securities Account', 'CD Account', 'Online', 'CreditCard']
```

```python
UniversalBank[charc_col] = UniversalBank[charc_col].astype('category')
```

```python
UniversalBank = pd.get_dummies(columns=["Family","Education","Securities Account","CD Account","Online","CreditCard"],data=UniversalBank,drop_first=True)
```

```python
Y = UniversalBank.loc[:,"Personal Loan"]
Y.head(5)
```

```python
X = UniversalBank.drop(columns=["Personal Loan"],axis = 1)
X.head(5)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.20,random_state = 11)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train.iloc[:,:5])
X_train.iloc[:,:5] = scaler.transform(X_train.iloc[:,:5])
X_test.iloc[:,:5] = scaler.transform(X_test.iloc[:,:5])
```

```python
Logistic = LogisticRegression(class_weight={0:0.1,1:0.9},random_state=666)
```

```python
Logistic.fit(X_train,Y_train)
```

```python
y_pred_LR = Logistic.predict(X_test)
```

```python
y_pred_LR
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,y_pred_LR)
```
